```python
import numpy as np
import numpy.linalg as al
import matplotlib.pyplot as plt
import pandas as pd
import requests
import zipfile
import io

# Étape 1 : Télécharger le fichier ZIP depuis l’URL
urlpop = "https://www.insee.fr/fr/statistiques/fichier/2521169/base_cc_comparateur_csv.zip"
response = requests.get(urlpop)

# Étape 2 : Ouvrir le fichier ZIP en mémoire
with zipfile.ZipFile(io.BytesIO(response.content)) as z:
    print("Fichiers dans le ZIP :")
    print(z.namelist())  # Liste des fichiers dans le ZIP

    # Étape 3 : Lire directement un des fichiers CSV dans le ZIP
    # ⚠️ Vérifie le nom exact dans la liste imprimée ci-dessus
    with z.open("base_cc_comparateur.csv") as csvfile:
        df = pd.read_csv(csvfile, sep=";", low_memory=False)

# Vérification
print(df.shape)
print(df.head())
```
